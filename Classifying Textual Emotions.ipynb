# Train Dataset

```python
label_list = [] 
sentence_list = []
with open('/content/drive/MyDrive/Colab Notebooks/dataset/train.txt','r') as dataset:
  for train_dataset in dataset.readlines():
      lineinfo=train_dataset.split('\"', 2)
      sentence_list.append(lineinfo[2].strip())
      label_list.append(lineinfo[1]) 
      
print(label_list)
```

```python
print(sentence_list)
```

```python
#using dictionary
unique_labels = list(set(label_list))
label_dic = {}
for indx, val in enumerate(unique_labels):
  label_dic[val] = indx
print(label_dic)
```

```python
# labels = []
# for val in label_list:
#   labels.append(label_dic[val])

labels = [label_dic[val] for val in label_list]
labels
```

```python
# tokenization and converting words into sequences
tokenizer = Tokenizer()
# tokenizer.fit_on_texts(label_list) 
tokenizer.fit_on_texts(sentence_list)
# dense_train_out = tokenizer.texts_to_sequences(label_list)
dense_train_in = tokenizer.texts_to_sequences(sentence_list)

print(np.array(label_list))
print(np.array(labels))
```

```python
# padding the training documents in order to make them equal length
MAX_LENGTH = 70

padded_train_ex = pad_sequences(dense_train_in, maxlen=MAX_LENGTH, padding='post')

for pd_sen in padded_train_ex:
  print(pd_sen)
```

```python
# VOCAB_SIZE = 17267
temp=0
for line in dense_train_in:
  for word in line:
    temp=max(temp,word)
     
    print(temp)
```

```python
# Model Declaration
VOCAB_SIZE = 17262

model = Sequential()

# Embedding Layer
embedding_layer = Embedding(input_dim=VOCAB_SIZE, output_dim=6, input_length=MAX_LENGTH)
model.add(embedding_layer)

# Bidirectional LSTM
forward_layers = LSTM(units=100, return_sequences=False)
backward_layers = LSTM(units=100, return_sequences=False, go_backwards=True)
model.add(Bidirectional(layer=forward_layers, backward_layer=backward_layers))

# Output Layer
model.add(Dense(units=100, activation='sigmoid'))
model.add(Dense(units=70, activation='sigmoid'))
model.add(Dense(units=50, activation='sigmoid'))
model.add(Dense(units=30, activation='sigmoid'))
model.add(Dense(units=10, activation='sigmoid'))
model.add(Dense(units=2, activation='sigmoid'))

model.compile(optimizer='adam', loss='Poisson', metrics=['msle'])

print(model.summary())
```

```python
labels = np.array(labels)
```

```python
# tf.convert_to_tensor(p)
model.fit((padded_train_ex),(labels), epochs=10, verbose=1)
# model.fit(p, q, epochs=3, validation_data=(x_test, y_test))
# model.fit((padded_train_ex), np.array(dense_train_out), epochs=1, verbose=1)
```
